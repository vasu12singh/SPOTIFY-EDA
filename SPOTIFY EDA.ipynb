### Correlation heatmap between variables¶

```python
# For sampling data
sample_df=df_track.sample(int(0.005*len(df_track)))
len(sample_df)
```

### Loudness vs Energy Correlation

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="loudness",x="energy",color="c",marker="*").set(title="Loudness vs Energy corr")
```

### Popularity vs Accoustic Correlation

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="popularity",x="acousticness",color="r",marker="*").set(title="popularity vs accoustic corr")
```

```python
df_track["release_date"]=pd.to_datetime(df_track["release_date"])
df_track["year"]=df_track["release_date"].dt.year
```

### Number of Songs per year

```python
sns.displot(df_track["year"],discrete=True,aspect=2,height=5,kind="hist").set(title="Number of song per year")
```

### Year vs Duration Barplot

```python
total_dr=df_track["duration"]
sns.set_style(style="whitegrid")
fig_dims = (16,6)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=df_track["year"],y=total_dr,ax=ax,errwidth=False).set(title="Year vs Duration")
```

### Year vs Duration Lineplot

```python
sns.set_style(style="whitegrid")
fig_dims=(10,5)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=df_track["year"],y=total_dr,ax=ax).set(title="year vs duration")
plt.xticks(rotation=60)
```
